##### MISNT로 다시 깊게 들여다 보자

```python
import tensorflow as tf

mnist = tf.keras.datasets.mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train, X_test = X_train/255, X_test/255

X_train = X_train.reshape((60000,28,28,1))
X_test = X_test.reshape((10000,28,28,1))
```

- 데이터를 255로 나눈 이유

픽셀이 255가 최대값이라 255까지만 데이터를 짜른것이다.

- 데이터뒤에 차원을 1로 reshape 한 이유

차원을 뒤에 1로 꼭 해줘야한다. -> 채널때문에

하나의 이미지를 필터로 여러번 거쳐 할건데 그러면 차원을 3차원으로 표시해야 해서이다.


ex) 만약에 28x28의 이미지를 32채널에 거친다 = 32개의 이미지 특성을 찾겠다( = 32개의 필터를 거쳐 28x28x32개가 나온다)

이미지의 특징을 추출한 부분을 32개로 세분화 시켰다고 이해하면 된다.

따라서 뒤에 3차원 계산이 필요하기 때문에 이미지 뒤에 1을 추가해 3차원으로 만들어 줘야한다.

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(3, kernel_size=(3,3), strides=(1,1),
                padding='same', activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D(pool_size=(2,2), strides=(2,2)),
    #layers.Conv2D(64,(2,2), activation='relu', padding='same'),
    #layers.MaxPooling2D(pool_size=(2,2)),
    layers.Dropout(0.25),
    layers.Flatten(),
    layers.Dense(1000, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.summary()
```

- 첫번째 채널은 3개
- stride는 필터의 스텝, 즉 1칸씩 필터 이동
- padding : 필터와 이미지가 합성곱을 지나면 차원이 축소되는데 축소를 어떻게 할지 결정
- 활성화 함수 : relu

##### 모델에서 아직 학습하지 않은 레이어의 W를 보자

```python
model.layers
```

##### W의 평균과 표준

```python
conv = model.layers[0]
conv_weights = conv.weights[0].numpy()
conv_weights.mean(), conv_weights.std()
```

##### W의 분포

```python
import matplotlib.pyplot as plt

plt.hist(conv_weights.reshape(-1,1))
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```

##### 컬볼루션 레이어의 필터(커널) (3x3 크기)

```python
fig, ax = plt.subplots(1,3, figsize=(15,5))
for i in range(3):
    ax[i].imshow(conv_weights[:,:,0,i], vmin = -0.5, vmax = 0.5)
    ax[i].axis('off')
    
plt.show()
```

- vmin, vmax란

필터 안에 랜덤값이 적용될텐데 동시에 비교할라고 최대 최소 스케일을 적용해 비교하기 위함이다.

##### 학습하자

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

hist = model.fit(X_train, y_train, epochs=5, verbose=1,
                validation_data=(X_test, y_test))
```

##### 학습 후의 필터의 변화를 살펴보자

```python
conv = model.layers[0]
conv_weights = conv.weights[0].numpy()

fig, ax = plt.subplots(1,3, figsize=(15,5))
for i in range(3):
    ax[i].imshow(conv_weights[:,:,0,i], vmin = -0.5, vmax = 0.5)
    ax[i].axis('off')
    
plt.show()
```

##### 5로 한번 살펴보자

```python
plt.imshow(X_train[0], cmap='gray');
```

##### 5이미지를 컨벌루션 레이어에 통과시킨 출력값을 뽑자
##### 먼저 컨벌루션 레이어만 추출

```python
## 컨벌루션 레이어만 따로 추출 ( 이름 : conv_layer_inout)
conv_layer_inout = tf.keras.Model(model.input, model.layers[0].output)
conv_layer_inout.summary()
```

##### 위에서 추출한 컨벌루션 레이어에 input데이터를 넣어보자

```python
inputs = X_train[0].reshape(-1, 28, 28, 1)

feature_maps = conv_layer_inout.predict(inputs)

feature_maps.shape, feature_maps[0,:,:,0].shape
```

- fit은 이미 했고 W는 고정이 되어있는 상태, 거기서 컨벌루션 레이어만 불러와서 보는 중
- 위에서 3x3 짜리 필터가 3개 지정해주었다. 그래서 결과는 (1x28x28)가 3개 있다는 소리다.
- 필터를 거친 5를 출력해보자

```python
fig, ax = plt.subplots(1,3, figsize=(15,5))
for i in range(3):
    ax[i].imshow(feature_maps[0,:,:,i])
    ax[i].axis('off')
    
plt.show()
```

- 컨벌루션레이어에서는 이렇게 5의 특성을 파악했구나~ 라고 알수있는 것이다.

##### 위 작업을 함수로 만들어 반복작업해보자

```python
def draw_feature_maps(n):
    inputs = X_train[n].reshape(-1, 28, 28, 1)

    feature_maps = conv_layer_inout.predict(inputs)
    
    fig, ax = plt.subplots(1,4, figsize=(15,5))
    
    ax[0].imshow(inputs[0, :, :, 0], cmap='gray'); # 원본 사진
    
    for i in range(1,4):
        ax[i].imshow(feature_maps[0,:,:,i-1]) # 필터 거친 사진
        ax[i].axis('off')
    
    plt.show()
```

```python
draw_feature_maps(1)
```

```python
draw_feature_maps(14)
```

```python
draw_feature_maps(54)
```

##### 채널을 증가시켜서 비교해보자!! ( 3 -> 8 )

```python
from tensorflow.keras import layers, models

model1 = models.Sequential([
    layers.Conv2D(8, kernel_size=(3,3), strides=(1,1),
                padding='same', activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D(pool_size=(2,2), strides=(2,2)),
    #layers.Conv2D(64,(2,2), activation='relu', padding='same'),
    #layers.MaxPooling2D(pool_size=(2,2)),
    layers.Dropout(0.25),
    layers.Flatten(),
    layers.Dense(1000, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model1.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

hist1 = model1.fit(X_train, y_train, epochs=5, verbose=1,
                validation_data=(X_test, y_test))
```

```python
## 컨벌루션 레이어만 따로 추출 ( 이름 : conv_layer_inout)
conv_layer_inout1 = tf.keras.Model(model1.input, model1.layers[0].output)

## 이미지 비교 출력
def draw_feature_maps1(n):
    inputs = X_train[n].reshape(-1, 28, 28, 1)

    feature_maps1 = conv_layer_inout1.predict(inputs)
    
    fig, ax = plt.subplots(1,9, figsize=(15,5)) # 채널 증가로 4 -> 9
    
    ax[0].imshow(inputs[0, :, :, 0], cmap='gray'); # 원본 사진
    
    for i in range(1,9): # 채널 증가로 4 -> 9
        ax[i].imshow(feature_maps1[0,:,:,i-1]) # 필터 거친 사진
        ax[i].axis('off')
    
    plt.show()
```

```python
draw_feature_maps1(1)
```

```python
draw_feature_maps1(14)
```

```python
draw_feature_maps1(54)
```
