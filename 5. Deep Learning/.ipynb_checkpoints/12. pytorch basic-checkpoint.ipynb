```python
## 더하기

x = torch.tensor(3.5)

print(x+6.2)
```

```python
## 기울기 계산 할수 있게 하기 -> x를 어떤 함수에 넣으면 기울기 반환 가능

x = torch.tensor(3.5, requires_grad=True)

print(x)
```

```python
## x가 tensor타입이기 때문에 자동으로 y도 tensor타입

y = (x-1) * (x-2) * (x-3) # x = 3.5일때의 y 값

print(y)
```

```python
# 기울기 계산하기

y.backward()
x.grad
```

### 보스톤 선형회귀 torch학습

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
columns =['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dist', 'rad', 'tax', 'ptratio', 'black', 'lstat']

raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

boston_pd = pd.DataFrame(data,columns=columns)
boston_pd['target(price)'] = target

boston_pd.head()
```

```python
import torch
import torch.nn as nn            # 뉴럴 넷
import torch.nn.functional as F  # 함수
import torch.optim as optim      # 최적화
```

```python
## 학습에 필요한(선정한) 특성 

cols = ['target(price)', 'indus', 'rm', 'lstat', 'nox', 'dist']
data = torch.from_numpy(boston_pd[cols].values).float() # 넘파이 -> 토치.float 자료형으로 (많이사용함!)

data.shape
```

```python
### 특성과 라벨로 분리
y = data[:, :1]
x = data[:,1:]

print(x.shape, y.shape)
```

```python
## 하이퍼 파라미터
n_epochs = 2000
learning_rate = 1e-3
print_interval = 100 # 몇번마다 print
```

```python
## linear 모델 정의
model = nn.Linear(x.size(-1), y.size(-1))
model
```

```python
## 최적화 방법
optimizer = optim.SGD(model.parameters(), lr=learning_rate)
```

```python
## 학습 시작

for i in range(n_epochs):
    y_hat = model(x)
    loss = F.mse_loss(y_hat, y) # loss를 계산해라
    
    optimizer.zero_grad()# gard를 0로 셋팅해라
    # -> 가중치를 업데이트하는 재료의 gard를 0으로 잡는 것(이유 : 반복문이니까 이전 값이남아있으면 안되서)
    loss.backward() # 역전파
    
    optimizer.step() # optimizer를 실행해라
    
    if (i + 1)% print_interval == 0:
        print('Epoch %d : loss = %.4e' % (i + 1, loss))
```

```python
## 학습 결과

df = pd.DataFrame(torch.cat([y, y_hat], dim=1).detach_().numpy(), columns=['y','y_hat'])

sns.pairplot(df, height=5)
plt.show()
```

- 왼쪽 위 막대그래프가 참
- 참 막대그래프의 오른쪽에 막대가 예측에서 사라짐
    - 이상치여서 버렸다 or 리니어로 학습하기 어려웠다. 라고 볼수 있다.

## 유방암 예측

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['class'] = cancer.target

df.tail()
```

```python
cols = ['mean radius', 'mean texture',
        'mean smoothness', 'mean compactness', 'mean concave points',
        'worst radius', 'worst texture',
        'worst smoothness', 'worst compactness', 'worst concave points',
        'class'
       ]
```

```python
for c in cols[:-1]:
    sns.histplot(df, x=c, hue=cols[-1], bins=50, stat='probability')
    plt.show()
```

```python
import torch
import torch.nn as nn            # 뉴럴 넷
import torch.nn.functional as F  # 함수
import torch.optim as optim      # 최적화
```

```python
## 학습에 필요한(선정한) 특성 

data = torch.from_numpy(df[cols].values).float() # 넘파이 -> 토치.float 자료형으로 (많이사용함!)
data.shape
```

```python
### 특성과 라벨로 분리

x = data[:, :-1]
y = data[:, -1:]

print(x.shape, y.shape)
```

```python
## 하이퍼 파라미터
n_epochs = 200000
learning_rate = 1e-2
print_interval = 10000 # 몇번마다 print
```

```python
## 클래스로 모델 선언

class Mymodel(nn.Module):
    
    def __init__(self, input_dim, output_dim):
        self.input_dim = input_dim
        self.output_dim = output_dim

        super().__init__() # super로 init을 받아줘야만 module의 속성을 상속받는다?
        
        self.linear = nn.Linear(input_dim, output_dim)
        self.act = nn.Sigmoid() # 출력단 : 이진 분류
        
    def forward(self, x):
        # x = (batch_size, input_dim)
        y = self.act(self.linear(x))
        # y = (batch_size, output_dim)
        
        return y
```

## 모델, loss, optimizer 선언

```python
model = Mymodel(input_dim=x.size(-1),
               output_dim=y.size(-1))

crit = nn.BCELoss()  # binary crossentropy

optimizer = optim.SGD(model.parameters(), lr=learning_rate)
```

```python
## 학습 시작

for i in range(n_epochs):
    y_hat = model(x)
    loss = crit(y_hat, y) # loss를 계산해라
    
    optimizer.zero_grad()# gard를 0로 셋팅해라
    # -> 가중치를 업데이트하는 재료의 gard를 0으로 잡는 것(이유 : 반복문이니까 이전 값이남아있으면 안되서)
    loss.backward() # 역전파
    
    optimizer.step() # optimizer를 실행해라
    
    if (i + 1)% print_interval == 0:
        print('Epoch %d : loss = %.4e' % (i + 1, loss))
```

```python
## 학습 결과

df_result = pd.DataFrame(torch.cat([y, y_hat], dim=1).detach_().numpy(), columns=['y','y_hat'])

sns.histplot(df_result, x='y_hat', hue='y', bins=50, stat='probability')
plt.show()
```
