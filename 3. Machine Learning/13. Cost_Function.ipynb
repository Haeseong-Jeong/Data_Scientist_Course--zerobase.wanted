poly1d : 다항식을 정의하고 조작할 수 있는 강력한 도구

```python
import numpy as np
```

```python
a = np.poly1d([1, 1]) # x+1
b = np.poly1d([1, -1]) # x-1
```

```python
a, b
```

```python
a*b # (x+1)(x-1) = x^2 - 1
```

## poly1d를 사용해서 직접 Cost Function을 구해보자


```python
np.poly1d([2, -1])**2 + np.poly1d([3, -5])**2 + np.poly1d([5, -6])**2
```
