## 데이터 생성

````python
import pandas as pd

data = {'x' : [1,2,3,4,5], 'y': [1,3,4,6,5]}
df = pd.DataFrame(data)
df
````

## 가설 세워주기 : $y = ax + b$

````python
import statsmodels.formula.api as smf

lm_model = smf.ols(formula="y ~ x", data=df).fit()
````

## 결과 : params

````python
lm_model.params
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
sns.lmplot(x='x',y='y',data=df, );
plt.grid()
plt.xlim([0,7])
plt.ylim([0,7])
````

## 잔차 평가 residue
잔차는 평균이 0인 정규분포를 따르는 것 이어야 한대.. 왜?

````python
resid = lm_model.resid
resid
````

## 결정계수 R-Squared


```
(예측한 선과 평균 선의 오차 제곱)
----------------------------------
 (에러 평균선과 참값의 오차 제곱)
```

## numpy로 직접 결정계수 계산하기


````python
import numpy as np

mu = np.mean(df.y)
y=df.y
yhat = lm_model.predict()
R_squared = np.sum((yhat - mu)**2 / np.sum((y - mu)**2))
R_squared
````

````python
# 간단하게 
lm_model.rsquared
````
