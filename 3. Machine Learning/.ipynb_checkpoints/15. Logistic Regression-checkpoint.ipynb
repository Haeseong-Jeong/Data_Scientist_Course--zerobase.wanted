### 시그모이드 함수 직접 그려보기

```python
import numpy as np

z = np.arange(-10, 10, 0.01)
g = 1 / (1+np.exp(-z))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(z, g);
plt.grid()
```

```python
plt.figure(figsize=(12, 8))

# 사용자 지정할수 있는 그래프
ax = plt.gca()

ax.plot(z, g);
ax.spines['left'].set_position('zero') # y축을 가운데로 (left -> zero)
ax.spines['bottom'].set_position('center') # x 축을 가운데로 (bottom > center)

ax.spines['right'].set_color('None') # 위에 축 숨기기
ax.spines['top'].set_color('None') # 오른쪽 축 숨기기

plt.show()
```

## logistic Regression에서의 cost function 재정의

```python
h = np.arange(0.01, 1, 0.01)

C0 = -np.log(1-h)
C1 = -np.log(h)

plt.figure(figsize=(12, 8))
plt.plot(h, C0, label='y = 0')
plt.plot(h, C1, label='y = 1')
plt.legend()

plt.show()
```

```python
import pandas as pd

wine_url = "https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/wine.csv"

wine = pd.read_csv(wine_url, index_col=0)
wine.head()
```

```python
wine['taste'] = [1 if grade>5 else 0 for grade in wine['quality']]

X = wine.drop(['taste', 'quality'], axis=1)
y = wine['taste']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression(solver='liblinear', random_state=13) # solver : 최적화 알고리즘 고르기
lr.fit(X_train, y_train)

y_pred_tr = lr.predict(X_train)
y_pred_test = lr.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ',accuracy_score(y_test, y_pred_test))
```

## 스케일러 적용 및 파이프라인 구축

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

estimators = [
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(solver='liblinear', random_state=13))
]

pipe = Pipeline(estimators)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_tr = pipe.predict(X_train)
y_pred_test = pipe.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ',accuracy_score(y_test, y_pred_test))
```

- 효과는 미미하다!!!!..........

## Decision Tree와 Logistic regression 비교!!!

```python
from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)
```

```python
models = {
    'logistic regression' : pipe,
    'decision tree' : wine_tree
}
```

```python
from sklearn.metrics import roc_curve

plt.figure(figsize=(12,6))
plt.plot([0,1],[0,1], label = 'random_guess')

for models_name, model in models.items():
    pred = model.predict_proba(X_test)[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, pred)
    plt.plot(fpr, tpr, label=models_name)
    
plt.grid()
plt.legend()
plt.show()
```
