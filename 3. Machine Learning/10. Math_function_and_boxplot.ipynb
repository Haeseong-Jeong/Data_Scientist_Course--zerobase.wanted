```python
def log(x, base):
    return np.log(x)/np.log(base)

x1 = np.linspace(0.0001, 5, 1000)
x2 = np.linspace(0.01, 5, 100)

y11, y12 = log(x1, 10), log(x2, np.e)
y21, y22 = log(x1, 1/10), log(x2, 1/np.e)
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))

ax[0].plot(x1, y11, label=r"$\log_{10} x$", color='k')
ax[0].plot(x2, y12, '--', label=r"$\log_{e} x$", color='k')

ax[0].set_xlabel("$x$", fontsize=25)
ax[0].set_ylabel("$y$", fontsize=25)
ax[0].legend(fontsize=25, loc='lower right')

ax[1].plot(x1, y21, label=r"$\log_{1/10} x$", color='k')
ax[1].plot(x2, y22, '--', label=r"$\log_{1/e} x$", color='k')

ax[1].set_xlabel("$x$", fontsize=25)
ax[1].set_ylabel("$y$", fontsize=25)
ax[1].legend(fontsize=25, loc='upper right')

plt.show()
```

```python
u = np.linspace(0, 1, 30)
v = np.linspace(0, 1, 30)

U, V =np.meshgrid(u, v)

X = U
Y = V

Z = (1 + U**2) + V / (1 + V**2)
```

```python
fig = plt.figure(figsize=(7,7))
ax = plt.axes(projection='3d')

ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)

ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.set_zlabel('$z$', fontsize=20)


ax.scatter3D(X,Y,Z, marker='.', color='r')
```

```python
x = np.linspace(-4, 4, 100)

# y = f(x)
y = x**3 - 15*x + 30

# z = g(y)
z = np.log(y)
```

###  

```python
fig, ax = plt.subplots(1, 2, figsize=(12,6))

ax[0].plot(x, y, label=r"$x^3 - 15x + 30$", color='r')
ax[0].legend(fontsize=18)

ax[1].plot(y, z, label=r"$\log_(y)$", color='b')
ax[1].legend(fontsize=18)

plt.show()
```

###  

```python
fig, ax = plt.subplots(1, 2, figsize=(12,6))

ax[0].plot(x, z, '--', label=r"$\log_(f(x))$", color='r')
ax[0].legend(fontsize=18)

ax[1].plot(x, y, label=r"$x^3 - 15x + 30$", color='b')
ax[1].legend(fontsize=18)

ax_tmp = ax[1].twinx()
ax_tmp.plot(x, z, '--', label=r"$\log_(f(x))$", color='r')

plt.show()
```

## box plot 

-   : IQR
-   * 1.5     Outlier  

```python
samples = [1,7,9,16,36,39,45,45,46,48,51,100,101]
tmp_y = [1]*len(samples)
```

```python
len(samples), len(tmp_y)
```

```python
plt.figure(figsize=(12,4))
plt.scatter(samples, tmp_y)
plt.grid()
plt.show()
```

##   

```python
np.median(samples) # = np.percentile(samples,50)
```

```python
np.percentile(samples,25)
```

```python
np.percentile(samples,75)
```

```python
#   : IQR
np.percentile(samples,75) - np.percentile(samples,25)
```

```python
iqr = np.percentile(samples,75) - np.percentile(samples,25)
iqr * 1.5
```

```python
q1 = np.percentile(samples,25)
q2 = np.percentile(samples,50)
q3 = np.percentile(samples,75)
iqr = np.percentile(samples,75) - np.percentile(samples,25)

upper_fence = q3 + iqr*1.5
lower_fence = q1 - iqr*1.5
```

```python
plt.figure(figsize=(12,4))
plt.scatter(samples, tmp_y)
plt.axvline(x=q1, color='r')
plt.axvline(x=q2, color='g')
plt.axvline(x=q3, color='b')
plt.axvline(x=upper_fence, color='k', ls='--')
plt.axvline(x=lower_fence, color='k', ls='dashed')
plt.grid()
plt.show()
```
