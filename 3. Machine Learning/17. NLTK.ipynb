# 나이브 베이즈 분류

```python
from nltk.tokenize import word_tokenize
import nltk
```

#### 나이브 베이즈는 지도 학습 분류기라 답을 알려줘야한다.

```python
train = [
    ('i like you', 'pos'),
    ('i hate you', 'neg'),
    ('you like me', 'neg'),
    ('i like her', 'pos')
]
```

#### 말뭉치 만들기

```python
train[0]
```

```python
sentence = train[0]
word_tokenize(sentence[0])
```

```python
all_words = set(
    word.lower() for sentence in train for word in word_tokenize(sentence[0])
)
all_words
```

#### 말뭉치에 단어가 있다 없다를 표기한다.
- 단어를 독립적으로 보고 한 문장에서 그 단어가 있고 없고를 확인해서 감정이 긍정인지 부정인지를 판단하는지표로 세운다

```python
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1]) for x in train]
t
```

#### 나이브 베이즈 분류 학습시키기
- sklearn 에서는 fit
- 여기선 train

```python
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

### 새로운 문장으로 테스트해보자

```python
test_sentence = 'i like MeRui'
test_sent_features = {
    word.lower(): (word in word_tokenize(test_sentence.lower())) for word in all_words
}

test_sent_features
```

```python
classifier.classify(test_sent_features)
```

# 한글 나이브 베이즈

```python
from konlpy.tag import Okt

pos_tagger = Okt()
```

```python
train = [
    ("메리가 좋아", 'pos'),
    ("고양이도 좋아", 'pos'),
    ("난 수업이 지루해", 'neg'),
    ("메리는 이쁜 고양이야", 'pos'),
    ("난 마치고 메리랑 놀거야", 'pos')
]
```

## 토큰화 시키고 중복을 제거한 리스트가 all_words가 된다.

```python
from nltk.tokenize import word_tokenize
import nltk

all_words = set(
    word.lower() for sentence in train for word in word_tokenize(sentence[0])
)
all_words
```

```python
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1]) for x in train]
t
```

```python
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```python
test_sentence = '난 수업이 마치면 메리랑 놀거야'

test_sent_features = {
    word.lower(): (word in word_tokenize(test_sentence.lower())) for word in all_words
}

test_sent_features
```

```python
classifier.classify(test_sent_features)
```

## 품사를 붙이자(태그)

```python
def tokenize(doc):
    return ['/'.join(t) for t in pos_tagger.pos(doc, norm=True,stem=True)]
```

```python
train_docs = [(tokenize(row[0]), row[1]) for row in train]
train_docs
```

## 말뭉치를 만들자

```python
tokens = [t for d in train_docs for t in d[0]]
tokens
```

```python
def term_exists(doc):
    return {word: (word in set(doc)) for word in tokens}
```

```python
train_xy = [(term_exists(d), c) for d, c in train_docs]
train_xy
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_xy)
classifier.show_most_informative_features()
```

```python
test_sentence = [('난 수업이 마치면 메리랑 놀거야')]

test_docs = pos_tagger.pos(test_sentence[0])
test_docs
```

```python
test_sent_features = { word: (word in tokens) for word in test_docs}

test_sent_features
```

```python
classifier.classify(test_sent_features)
```

# 문서 벡터화

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df=1)
```

```python
contents =[
    '상처받은 아이들은 너무 일찍 커버려',
    '내가 상처받은 거 아는 사람 불편해',
    '잘 사는 사람들은 좋은 사람 되기 쉬워',
    '아무 일도 아니야 괜찮아'    
]
```

```python
from konlpy.tag import Okt

t = Okt()
```

```python
contents_tokens = [ t.morphs(row) for row in contents ]
contents_tokens
```

nltk를 쓸거다 -> nltk는 한 문장이 들어오는게 좋다 -> CountVectorizer는 띄어쓰기로 구분한다.-> 그래서 띄어쓰기를 붙여서 한 문장으로 만들어줘야한다.

```python
contents_for_vectorize = []

for content in contents_tokens: # ['상처', '받은', '아이',]
    sentence = ""
    for word in content:
        sentence = sentence + ' ' + word # '상처 ' + '받은 ' +'아이 ' 
        
    contents_for_vectorize.append(sentence)
contents_for_vectorize
```

```python
X = vectorizer.fit_transform(contents_for_vectorize)
X
```

```python
X.shape
```

```python
vectorizer.get_feature_names_out()
```

```python
len(vectorizer.get_feature_names_out())
```

```python
new_post = ['상처받기 싫어 괜찮아']

new_post_tokens = [ t.morphs(row) for row in new_post ]

new_post_for_vectorize = []

for content in new_post_tokens: # ['상처', '받은', '아이',]
    sentence = ""
    for word in content:
        sentence = sentence + ' ' + word # '상처 ' + '받은 ' +'아이 ' 
        
    new_post_for_vectorize.append(sentence)
    
new_post_for_vectorize
```

```python
new_post_vec = vectorizer.transform(new_post_for_vectorize)
new_post_vec.toarray()
```

```python
import scipy as sp

def dist_raw(v1, v2):
    delta = v1-v2
    return sp.linalg.norm(delta.toarray())
```

```python
dist = [ dist_raw(each, new_post_vec) for each in X]
dist
```

```python
print('Best post is ', dist.index(min(dist)), ', dist = ', min(dist))
print('Test post is ', new_post)
print('Best dist post is ', contents[dist.index(min(dist))])
```

# TF-IDF 

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df=1, decode_error='ignore')
```

```python
X = vectorizer.fit_transform(contents_for_vectorize)
X
```

```python
X.shape
```

```python
X.toarray().transpose()
```

```python
new_post_vec = vectorizer.transform(new_post_for_vectorize)
new_post_vec.toarray()
```

```python
import scipy as sp

def dist_raw(v1, v2):
    v1_normalized = v1 / sp.linalg.norm(v1.toarray())
    v2_normalized = v2 / sp.linalg.norm(v2.toarray())
    
    delta = v1_normalized - v2_normalized
    return sp.linalg.norm(delta.toarray())
```

```python
dist = [ dist_raw(each, new_post_vec) for each in X]
dist
```

```python
print('Best post is ', dist.index(min(dist)), ', dist = ', min(dist))
print('Test post is ', new_post)
print('Best dist post is ', contents[dist.index(min(dist))])
```
