````python
from pandas_datareader import data
from prophet import Prophet
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
time = np.linspace(0,1,365*2)
result = np.sin(2*np.pi*12*time)
ds = pd.date_range("2018-01-01", periods=365*2, freq = "D")
df = pd.DataFrame({"ds":ds, "y": result})
df.head()
````

````python
df["y"].plot(figsize=(10,6));
````

````python
m = Prophet(yearly_seasonality = True, daily_seasonality=True);
m.fit(df);
````

````python
future = m.make_future_dataframe(periods=30);
forecast = m.predict(future);
````

````python
m.plot(forecast);
## ÏãúÍ∞ÅÌôîÎ•º ÏßÑÌñâ Ìï† Í≤ΩÏö∞ Ï∞®Ìä∏Í∞Ä ÎëêÍ∞úÎ°ú ÎÇòÏò§Îäî Î≤ÑÍ∑∏Í∞Ä ÏûàÏäµÎãàÎã§. Îß® Îí§Ïóê ÏÑ∏ÎØ∏ÏΩúÎ°†ÏùÑ Î∂ôÏó¨Ï£ºÏÑ∏Ïöî
````

## ÏòàÏãú 2

````python
time = np.linspace(0,1,365*2)
result = np.sin(2*np.pi*12*time) + time

ds = pd.date_range("2018-01-01", periods=365*2, freq = "D")
df = pd.DataFrame({"ds":ds, "y": result})

df["y"].plot(figsize=(10,6));
````

````python
m = Prophet(yearly_seasonality=True, daily_seasonality=True) # 1. Í≥ÑÏ†àÏÑ±ÌïôÏäµÎ™®Îç∏ ÏÉùÏÑ±
m.fit(df) # 2. ÌïôÏäµÎ™®Îç∏Ïóê ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞ Ï∂îÍ∞Ä.
future = m.make_future_dataframe(periods=30) # 3. ÏõêÌïòÎäî ÎØ∏Îûò Íµ¨Í∞Ñ ÏÑ§Ï†ï
forecast = m.predict(future) # 4. ÎØ∏ÎûòÏùò Íµ¨Í∞Ñ ÏòàÏ∏° ÏãúÌÇ§Í∏∞
m.plot(forecast);
````

## ÏòàÏãú 3

````python
time = np.linspace(0,1,365*2)
result = np.sin(2*np.pi*12*time) + time + np.random.randn(365*2)/4

ds = pd.date_range("2018-01-01", periods=365*2, freq = "D")
df = pd.DataFrame({"ds":ds, "y": result})

df["y"].plot(figsize=(10,6));
````

````python
m = Prophet(yearly_seasonality=True, daily_seasonality=True)
m.fit(df)
future = m.make_future_dataframe(periods=30)
forecast = m.predict(future)
m.plot(forecast);
````
